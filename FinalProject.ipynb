```python
# ================================
# FINAL AQI PROJECT - Complete Code
# Comparative Analysis (2019–2025)
# Author: generated template (replace placeholders with your files/values)
# ================================

# -----------------------
# STEP 0: Install Required Libraries (if needed in Colab)
# -----------------------
!pip install pandas numpy matplotlib seaborn plotly scikit-learn statsmodels requests openpyxl kaleido

# -------------------------------
# STEP 1: Import Libraries
# -------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neural_network import MLPRegressor
import statsmodels.api as sm
import requests
import os
import io
import warnings
from datetime import timedelta
from sklearn.preprocessing import StandardScaler

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
RANDOM_SEED = 42

# -----------------------
# STEP 2: User config (EDIT THESE if needed)
# -----------------------
# 1) Paths to CPCB data CSV files (downloaded manually from CPCB).
#    It must have a date column and pollutant columns (PM2.5, PM10, etc).
#    Provide a dictionary where keys are city names and values are lists of paths.
#    NOTE: Ensure these paths are correct. If you uploaded files to Colab,
#    they might be in /content/drive/MyDrive/ if you mounted Drive, or
#    check the file browser for the correct /content/ path.
CPCB_CITY_PATHS = {
    "Bengaluru": [
        "/content/City_wise_raw_data_1Day_2019_Bengaluru_1Day.csv",
        "/content/City_wise_raw_data_1Day_2020_Bengaluru_1Day.csv",
        "/content/City_wise_raw_data_1Day_2021_Bengaluru_1Day.csv",
        "/content/City_wise_raw_data_1Day_2022_Bengaluru_1Day.csv",
        "/content/City_wise_raw_data_1Day_2023_Bengaluru_1Day.csv",
        "/content/City_wise_raw_data_1Day_2024_Bengaluru_1Day.csv"
    ],
    "Delhi": [
        "/content/City_wise_raw_data_1Day_2019_Delhi_1Day.csv",
        "/content/City_wise_raw_data_1Day_2020_Delhi_1Day.csv",
        "/content/City_wise_raw_data_1Day_2021_Delhi_1Day.csv",
        "/content/City_wise_raw_data_1Day_2022_Delhi_1Day.csv",
        "/content/City_wise_raw_data_1Day_2023_Delhi_1Day.csv",
        "/content/City_wise_raw_data_1Day_2024_Delhi_1Day.csv"
    ],
    "Mumbai": [
        "/content/City_wise_raw_data_1Day_2019_Mumbai_1Day.csv",
        "/content/City_wise_raw_data_1Day_2020_Mumbai_1Day.csv",
        "/content/City_wise_raw_data_1Day_2021_Mumbai_1Day.csv",
        "/content/City_wise_raw_data_1Day_2022_Mumbai_1Day.csv",
        "/content/City_wise_raw_data_1Day_2023_Mumbai_1Day.csv",
        "/content/City_wise_raw_data_1Day_2024_Mumbai_1Day.csv"
    ]
} # <-- Add more cities here if you have data

# 2) Column names in CPCB CSV (adjust if your file uses different names)
DATE_COL = "Timestamp"          # name of the date column in CPCB CSV (Updated based on data)
PM25_COL = "PM2.5"         # name for PM2.5 column
PM10_COL = "PM10"          # name for PM10 column
AQI_COL = "AQI"            # optional: AQI column name if available

# 3) City coordinates for NASA POWER (if you want to fetch weather via API) - For a single city or average
# Note: For multi-city analysis, ideally you'd have weather data per city. This fetches for a single point.
CITY_LAT = 20.5937   # Example: Center of India or specific city
CITY_LON = 78.9629   # Example: Center of India or specific city


# 4) NASA POWER fetch toggle: set True to fetch daily meteorology from NASA POWER
FETCH_NASA_POWER = True # Note: NASA POWER fetch is for a single point, not easily adapted for multiple cities globally without looping or averaging. Current code fetches for CITY_LAT/LON and applies to the merged dataset.

# 5) Date range for NASA POWER (YYYYMMDD)
NASA_START = "20190101"
NASA_END = "20251231"

# 6) Output file names
CLEANED_CSV_OUT = "cleaned_merged_dataset_multi_city.csv" # Changed output name for clarity
PLOTS_DIR = "plots"
os.makedirs(PLOTS_DIR, exist_ok=True)

# -----------------------
# STEP 3: Load & Combine CPCB data (Modified for multiple cities)
# -----------------------
print("Loading and combining CPCB data for cities:", list(CPCB_CITY_PATHS.keys()))
print("CPCB_CITY_PATHS dictionary:", CPCB_CITY_PATHS) # Debug print: print the dictionary
all_city_data = []

for city, cpcb_paths in CPCB_CITY_PATHS.items():
    print(f"\nProcessing data for {city}...")
    city_aqi_data = []
    for cpcb_path in cpcb_paths:
        print(f"Checking file existence: {cpcb_path}") # Debug print: check file path
        if not os.path.exists(cpcb_path):
            print(f"Warning: CPCB file not found at {cpcb_path}. Skipping this file for {city}.")
            continue

        try:
            # Try reading as CSV first
            df = pd.read_csv(cpcb_path)
            print(f"Successfully read {cpcb_path} for {city} as CSV. Shape: {df.shape}") # Debug print
        except Exception as e_csv:
            try:
                # If CSV fails, try Excel
                df = pd.read_excel(cpcb_path)
                print(f"Successfully read {cpcb_path} for {city} as Excel. Shape: {df.shape}") # Debug print
            except Exception as e_excel:
                print(f"Error reading {cpcb_path} for {city}: {e_csv if e_csv else e_excel}. Skipping this file.")
                continue
        city_aqi_data.append(df)

    if not city_aqi_data:
        print(f"No valid CPCB files found for {city}. Skipping this city.")
        continue

    # Concatenate all dataframes for this city
    aqi_city = pd.concat(city_aqi_data, ignore_index=True)
    print(f"Combined initial data for {city}. Shape: {aqi_city.shape}") # Debug print


    # Inspect columns to help you adjust mapping (Optional - uncomment for debugging)
    # print(f"CPCB columns for {city} before processing:", list(aqi_city.columns)[:30])

    # Try to parse date column robustly
    current_date_col = None
    if DATE_COL in aqi_city.columns:
        current_date_col = DATE_COL
        # Use errors='coerce' to turn unparseable dates into NaT
        aqi_city[current_date_col] = pd.to_datetime(aqi_city[current_date_col], errors="coerce")
    else:
        # try to find a date-like column
        date_candidates = [c for c in aqi_city.columns if "date" in c.lower() or "datetime" in c.lower()]
        if date_candidates:
            current_date_col = date_candidates[0]
            # Use errors='coerce' to turn unparseable dates into NaT
            aqi_city[current_date_col] = pd.to_datetime(aqi_city[current_date_col], errors="coerce")
        else:
            print(f"Warning: No date column found for {city}. Skipping this city.")
            continue

    # Drop rows where date parsing failed (NaT values)
    original_rows = len(aqi_city)
    aqi_city = aqi_city.dropna(subset=[current_date_col])
    dropped_rows = original_rows - len(aqi_city)
    if dropped_rows > 0:
        print(f"Dropped {dropped_rows} rows with invalid dates for {city}.") # Debug print

    # rename columns (if necessary) to standard names we'll use later
    cols = aqi_city.columns.tolist()
    rename_map = {}
    # Check for PM2.5
    if PM25_COL not in cols:
        for c in cols:
            if "pm2.5" in c.lower().replace(" ", "") or "pm25" in c.lower().replace(" ", ""):
                rename_map[c] = PM25_COL
                break
    # Check for PM10
    if PM10_COL not in cols:
        for c in cols:
            if "pm10" in c.lower().replace(" ", ""):
                rename_map[c] = PM10_COL
                break
    # Check for AQI
    if AQI_COL not in cols:
         for c in cols:
            if "aqi" in c.lower().replace(" ", ""):
                rename_map[c] = AQI_COL
                break

    if rename_map:
        aqi_city = aqi_city.rename(columns=rename_map)
        print(f"Renamed columns for {city}: {rename_map}") # Debug print

    # print(f"Using date column for {city}:", current_date_col)
    # print(f"After rename, columns for {city} include:", [c for c in [current_date_col, PM25_COL, PM10_COL, AQI_COL] if c in aqi_city.columns])

    # keep relevant columns (drop all others for initial processing)
    keep_cols = [current_date_col]
    for c in [PM25_COL, PM10_COL, AQI_COL]:
        if c in aqi_city.columns:
            keep_cols.append(c)

    # Ensure all keep_cols are actually in the dataframe before selecting
    actual_keep_cols = [c for c in keep_cols if c in aqi_city.columns]
    if not actual_keep_cols or current_date_col not in actual_keep_cols:
         print(f"Warning: No relevant pollutant or date columns found for {city}. Skipping this city.")
         continue

    aqi_city = aqi_city[actual_keep_cols].copy()
    print(f"DataFrame shape after keeping relevant columns for {city}: {aqi_city.shape}") # Debug print


    # Add City column BEFORE setting index
    aqi_city["City"] = city
    # Ensure the City column is included in the selection after adding it
    aqi_city = aqi_city[[current_date_col] + [col for col in actual_keep_cols if col != current_date_col] + ["City"]].copy()


    # set index and sort
    aqi_city = aqi_city.set_index(current_date_col).sort_index()


    # convert pollutant columns to numeric
    for c in [PM25_COL, PM10_COL, AQI_COL]:
        if c in aqi_city.columns:
            # Add debug print for numeric conversion
            original_dtype = aqi_city[c].dtype
            aqi_city[c] = pd.to_numeric(aqi_city[c], errors="coerce")
            if aqi_city[c].dtype != original_dtype:
                print(f"Converted column '{c}' for {city} to numeric.")
            if aqi_city[c].isnull().sum() > 0:
                print(f"Warning: {aqi_city[c].isnull().sum()} NaNs introduced during numeric conversion for column '{c}' in {city}.")


    # Aggregate to daily if needed (assuming raw data might be hourly/sub-daily)
    # For multi-city, apply aggregation *after* adding city column but *before* concat
    # ensure City column is kept during resampling
    # NOTE: When resampling multi-index, need to specify level or group by
    # We will drop the City index for resampling to date, then reset index to keep City column
    aqi_city_resampled = aqi_city.reset_index().groupby([aqi_city.index, "City"])[keep_cols[1:]].mean().reset_index(level=1)
    aqi_city_resampled.index.name = current_date_col

    all_city_data.append(aqi_city_resampled)
    print(f"Combined CPCB data loaded for {city}: rows =", len(aqi_city_resampled))
    print(f"Columns in resampled data for {city}: {list(aqi_city_resampled.columns)}") # Debug print


if not all_city_data:
    raise FileNotFoundError("No valid data found for any city at the specified paths.")

# Concatenate all city dataframes into one
merged_multi_city = pd.concat(all_city_data, ignore_index=False) # Keep original index (dates)
merged_multi_city.index.name = "date"

print("\nCombined data from all cities loaded:")
print("Shape:", merged_multi_city.shape)
print("Columns in merged_multi_city:", list(merged_multi_city.columns)) # Debug print
display(merged_multi_city.head())

# Re-assign the main merged dataframe for subsequent steps (Note: downstream steps might need modification for multi-city)
# merged = merged_multi_city.copy() # Removed this line as merged_multi_city is used directly now


# -----------------------
# STEP 4: Optional - Fetch NASA POWER daily meteorology (Note: Currently for a single point)
# -----------------------
# Note: NASA POWER fetch is for a single point, not easily adapted for multiple cities globally without looping or averaging.
# For multi-city analysis, ideally you'd have weather data per city.
nasa_df = None # Initialize nasa_df to None outside the if block

if FETCH_NASA_POWER:
    print("Fetching NASA POWER data (daily) for lat,lon:", CITY_LAT, CITY_LON)
    # We request daily time series for variables: T2M (temperature), RH2M (relative humidity), PRECTOTCORR (precip), WS2M (wind speed)
    # NASA POWER CSV API:
    url = (
        f"https://power.larc.nasa.gov/api/temporal/daily/point"
        f"?parameters=T2M,RH2M,PRECTOTCORR,WS2M"
        f"&start={NASA_START}&end={NASA_END}"
        f"&latitude={CITY_LAT}&longitude={CITY_LON}"
        f"&format=CSV"
    )
    r = requests.get(url)
    if r.status_code != 200:
        # Try fetching without PRECTOTCORR if it causes issues
        url = (
            f"https://power.larc.nasa.gov/api/temporal/daily/point"
            f"?parameters=T2M,RH2M,WS2M"
            f"&start={NASA_START}&end={NASA_END}"
            f"&latitude={CITY_LAT}&longitude={CITY_LON}"
            f"&format=CSV"
        )
        r = requests.get(url)
        if r.status_code != 200:
             print("NASA POWER API error: status code " + str(r.status_code))
             nasa_df = None
             print("Skipping NASA POWER fetch due to API error.")
        else:
            print("Fetched NASA POWER data without PRECTOTCORR.")

    if r.status_code == 200:
        # NASA returns a CSV with header lines - parse robustly
        content = r.content.decode("utf-8")
        # find the line number where the CSV header starts (look for 'YEAR,MO,DY' or 'DATE')
        lines = content.splitlines()
        header_line_idx = None
        for i, line in enumerate(lines[:30]):
            if line.strip().lower().startswith("year,mo,dy") or line.strip().lower().startswith("date,"):
                header_line_idx = i
                break
        if header_line_idx is None:
            # fallback: skip first 11 lines (typical)
            header_line_idx = 11
            print(f"Could not find header line, skipping first {header_line_idx} lines of NASA POWER data.")

        try:
            nasa_df = pd.read_csv(io.StringIO("\n".join(lines[header_line_idx:])))
            # handle two formats: DATE column or YEAR/MO/DY columns
            if "DATE" in nasa_df.columns:
                nasa_df["date"] = pd.to_datetime(nasa_df["DATE"], errors="coerce")
            else:
                if set(["YEAR","MO","DY"]).issubset(nasa_df.columns):
                    nasa_df["date"] = pd.to_datetime(nasa_df["YEAR"].astype(int).astype(str) + "-" +
                                                    nasa_df["MO"].astype(int).astype(str).str.zfill(2) + "-" +
                                                    nasa_df["DY"].astype(int).astype(str).str.zfill(2),
                                                    errors="coerce")
                else:
                    raise ValueError("NASA POWER: Could not parse date columns.")

            # Drop rows where date parsing failed
            nasa_df = nasa_df.dropna(subset=["date"])

            # select and rename variables to friendly names
            rename_nasa = {
                "T2M": "temp_C",
                "RH2M": "rh_percent",
                "PRECTOTCORR": "precip_mm",
                "WS2M": "wind_m_s"
            }
            # keep only those columns that exist and rename
            nasa_keep = ["date"] + [k for k in rename_nasa.keys() if k in nasa_df.columns]
            nasa_df = nasa_df[nasa_keep].rename(columns=rename_nasa)
            nasa_df = nasa_df.set_index("date").sort_index()

            # convert numeric
            for c in ["temp_C", "rh_percent", "precip_mm", "wind_m_s"]:
                if c in nasa_df.columns:
                    nasa_df[c] = pd.to_numeric(nasa_df[c], errors="coerce")

            print("NASA POWER data rows:", len(nasa_df))
            display(nasa_df.head())
        except Exception as e:
            print(f"Error processing NASA POWER data: {e}")
            nasa_df = None
            print("Skipping NASA POWER fetch due to processing error.")

else:
    nasa_df = None
    print("Skipping NASA POWER fetch. Set FETCH_NASA_POWER=True to fetch weather data.")

# -----------------------
# STEP 5: Merge datasets (CPCB + NASA) and basic cleaning
# -----------------------
# Note: Merging NASA POWER (single point) with multi-city data is for illustration.
# For robust multi-city weather analysis, you would need weather data per city.
# We will merge based on date, applying the same weather to all cities on a given day.
# This is a simplification for demonstration.
if nasa_df is not None and not merged_multi_city.empty:
    # Ensure both dataframes have a 'date' index before merging
    if not isinstance(merged_multi_city.index, pd.DatetimeIndex):
        merged_multi_city.index = pd.to_datetime(merged_multi_city.index, errors='coerce')
        merged_multi_city = merged_multi_city.dropna(subset=[merged_multi_city.index.name]) # Drop rows where date conversion failed
    if not isinstance(nasa_df.index, pd.DatetimeIndex):
        nasa_df.index = pd.to_datetime(nasa_df.index, errors='coerce')
        nasa_df = nasa_df.dropna(subset=[nasa_df.index.name])

    # Perform the merge using the date index
    # Keep the 'City' column which is part of the original multi-city structure
    # Use a left merge to keep all multi-city rows and add weather data where dates match
    merged_multi_city = merged_multi_city.merge(nasa_df, left_index=True, right_index=True, how='left')
    print("\nMerged AQI and NASA POWER data.")
    print("Merged data shape after adding weather:", merged_multi_city.shape) # Debug print


# Set a continuous daily index (helps interpolation / analysis) based on the range of the combined data
# This might not be needed anymore since we are working with the combined data directly
# and already have a date index from the CPCB data loading.
# Ensure the index is a DatetimeIndex for the next steps
if not isinstance(merged_multi_city.index, pd.DatetimeIndex):
    merged_multi_city.index = pd.to_datetime(merged_multi_city.index, errors='coerce')
    merged_multi_city = merged_multi_city.dropna(subset=[merged_multi_city.index.name]) # Drop rows where date conversion failed


# Impute missing pollutant/weather values
# Strategy: linear time interpolation within each city group
print("\nImputing missing values...")
# Identify all columns that are numeric and not 'City' to impute
cols_to_impute = merged_multi_city.select_dtypes(include=np.number).columns.tolist()
if 'year' in cols_to_impute: # 'year' is created later, exclude it if it exists already
    cols_to_impute.remove('year')


if cols_to_impute:
    print(f"Imputing columns: {cols_to_impute}")
    for col in cols_to_impute:
        if col in merged_multi_city.columns:
            # Add debug print before and after imputation
            nan_before = merged_multi_city[col].isnull().sum()
            merged_multi_city[col] = merged_multi_city.groupby("City")[col].transform(
                lambda x: x.interpolate(method="time", limit=14).fillna(method="ffill", limit=7).fillna(method="bfill", limit=7)
            )
            nan_after = merged_multi_city[col].isnull().sum()
            if nan_before > nan_after:
                 print(f"Imputed {nan_before - nan_after} missing values in column '{col}'. {nan_after} remain.") # Debug print
            elif nan_before > 0:
                 print(f"Warning: Could not impute all missing values in column '{col}'. {nan_before} remain.") # Debug print
        else:
             print(f"Warning: Column '{col}' not found for imputation.")
else:
    print("No numeric columns found for imputation.")


# Create additional time features
merged_multi_city["dayofyear"] = merged_multi_city.index.dayofyear
merged_multi_city["month"] = merged_multi_city.index.month
merged_multi_city["year"] = merged_multi_city.index.year
merged_multi_city["weekday"] = merged_multi_city.index.weekday

print("Merged dataset shape after cleaning and feature engineering:", merged_multi_city.shape)
display(merged_multi_city.head())

# Save initial cleaned merged CSV
merged_multi_city.to_csv(CLEANED_CSV_OUT)
print("Saved cleaned merged dataset to", CLEANED_CSV_OUT)


# -----------------------
# STEP 6: Exploratory Data Analysis (EDA) - Adapted for Multi-city
# -----------------------
print("\n--- Starting EDA (Step 6) ---") # Debug print
def timeseries_plot_multi_city(df, cols, title, save_as=None):
    """
    Generates a multi-city time series plot using Plotly Express.

    Args:
        df (pd.DataFrame): DataFrame with a MultiIndex ['date', 'City']
                           and pollutant columns.
        cols (list): List of column names to plot.
        title (str): Title of the plot.
        save_as (str, optional): Filename to save the plot. Saves to PLOTS_DIR.
                                 If None, plot is not saved.
    """
    print(f"Generating time series plot for {cols}...") # Debug print
    # Ensure df has 'date' as index and 'City' as a column or level
    # If index is date and City is a column, reset index for Plotly
    df_plot = df.copy()
    if 'City' in df_plot.columns and isinstance(df_plot.index, pd.DatetimeIndex):
         df_plot = df_plot.reset_index()
         print("Resetting index for Plotly plot.") # Debug print
    elif 'City' in df_plot.index.names and isinstance(df_plot.index.get_level_values('date'), pd.DatetimeIndex):
         df_plot = df_plot.reset_index() # Reset multi-index to get date and City as columns
         print("Resetting multi-index for Plotly plot.") # Debug print
    else:
         print("Warning: DataFrame structure not as expected for Plotly. Attempting to plot without resetting index.")
         df_plot['date'] = df_plot.index # Use index as date column if reset fails


    df_plot['date'] = pd.to_datetime(df_plot['date']) # Ensure date is datetime type
    print(f"DataFrame shape before plotting: {df_plot.shape}") # Debug print


    plot_data = []
    for col in cols:
        if col in df_plot.columns:
             # Debugging: Check for missing values in the column before plotting
             if df_plot[col].isnull().all():
                  print(f"Warning: Column '{col}' contains only missing values. Skipping plot for this column.")
                  continue
             elif df_plot[col].isnull().sum() > 0.5 * len(df_plot): # Warning if more than 50% missing
                  print(f"Warning: Column '{col}' has a large number of missing values ({df_plot[col].isnull().sum()} out of {len(df_plot)}). Plot may be sparse.")

            # Create a melted version for easier plotting of multiple pollutants
             melted_col_df = df_plot[['date', 'City', col]].melt(id_vars=['date', 'City'],
                                                                value_vars=[col],
                                                                var_name='Pollutant',
                                                                value_name='Concentration')
             # Drop rows where Concentration is NaN to avoid plotting gaps in Plotly
             melted_col_df = melted_col_df.dropna(subset=['Concentration'])
             plot_data.append(melted_col_df)
        else:
            print(f"Warning: Column '{col}' not found for time series plot.")

    if not plot_data:
        print("No valid columns found for time series plot after checking for missing data.")
        return

    plot_df = pd.concat(plot_data)

    # Check if plot_df is empty after dropping NaNs
    if plot_df.empty:
        print(f"Plotting data is empty after dropping missing values for columns {cols}. Cannot generate plot.")
        return

    print(f"Plotting {len(plot_df)} data points for columns {cols}") # Debug print

    fig = px.line(plot_df,
                  x='date',
                  y='Concentration',
                  color='City', # Color by City
                  line_dash='Pollutant', # Differentiate pollutants by line style
                  title=title,
                  labels={'date': 'Date', 'Concentration': 'Concentration (µg/m³)'}) # Add a generic y-label

    fig.update_layout(hovermode='x unified') # Improve hover experience

    fig.show() # Explicitly show the figure in the notebook output

    if save_as:
        # Ensure the plots directory exists
        if not os.path.exists(PLOTS_DIR):
             os.makedirs(PLOTS_DIR)

        png_path = os.path.join(PLOTS_DIR, save_as.replace('.html', '.png')) # Save as PNG
        html_path = os.path.join(PLOTS_DIR, save_as.replace('.png', '.html')) # Save as HTML

        try:
            fig.write_image(png_path)
            print(f"Saved plot to {png_path}")
        except Exception as e:
            print(f"Could not save plot as PNG. Error: {e}")
            print(f"Attempting to save as HTML instead.")
            try:
                fig.write_html(html_path)
                print(f"Saved plot as HTML to {html_path}")
            except Exception as html_e:
                print(f"Could not save plot as HTML either. Error: {html_e}")


# Time-series PM2.5 (Comparative)
ts_cols_multi = [c for c in [PM25_COL, PM10_COL] if c in merged_multi_city.columns]
if ts_cols_multi:
    print(f"\nAttempting to generate time series plots for {ts_cols_multi}...") # Debug print
    timeseries_plot_multi_city(merged_multi_city, [ts_cols_multi[0]], f"Daily Pollutant Levels ({ts_cols_multi[0]}) by City", save_as=f"ts_{ts_cols_multi[0].lower()}_multi_city.png")
    if len(ts_cols_multi) > 1:
         timeseries_plot_multi_city(merged_multi_city, [ts_cols_multi[1]], f"Daily Pollutant Levels ({ts_cols_multi[1]}) by City", save_as=f"ts_{ts_cols_multi[1].lower()}_multi_city.png")
else:
    print(f"Pollutant columns {PM25_COL} or {PM10_COL} not found for multi-city time-series plot.")

# Time-series AQI (Comparative) - Added based on user request
if AQI_COL in merged_multi_city.columns:
    print(f"\nAttempting to generate time series plot for {AQI_COL}...") # Debug print
    timeseries_plot_multi_city(merged_multi_city, [AQI_COL], f"Daily AQI Levels by City", save_as=f"ts_{AQI_COL.lower()}_multi_city.png")
else:
    print(f"Column {AQI_COL} not found for multi-city time-series plot.")


# Monthly seasonal boxplots by City
if PM25_COL in merged_multi_city.columns:
    print(f"\nGenerating monthly seasonal boxplot for {PM25_COL} by City...") # Debug print
    # Ensure 'month' is available for plotting - extract from index if needed
    df_plot_boxplot = merged_multi_city.copy().reset_index() # Reset index to make date and City columns
    df_plot_boxplot['month'] = df_plot_boxplot['date'].dt.month # Extract month from the date column

    print(f"Boxplot data shape for {PM25_COL}: {df_plot_boxplot.shape}") # Debug print
    if not df_plot_boxplot.empty:
        plt.figure(figsize=(14, 7)) # Adjusted figure size for multiple cities
        sns.boxplot(x='month', y=PM25_COL, hue='City', data=df_plot_boxplot) # Use hue for cities
        plt.xlabel("Month")
        plt.ylabel(PM25_COL)
        plt.title(f"Monthly Seasonal Boxplot ({PM25_COL}) by City")
        plt.legend(title="City", bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend
        plt.tight_layout()
        plt.savefig(os.path.join(PLOTS_DIR, f"boxplot_{PM25_COL.lower()}_by_month_city.png"), dpi=150)
        plt.show()
    else:
        print(f"Boxplot data for {PM25_COL} is empty. Cannot generate plot.")
else:
    print(f"Column {PM25_COL} not found for seasonal boxplot.")

# Yearly AQI comparison plots - Modified based on user request
if AQI_COL in merged_multi_city.columns and 'year' in merged_multi_city.columns:
    print(f"\nGenerating yearly average {AQI_COL} comparison plots by City...") # Debug print

    # Calculate yearly average AQI per city
    yearly_avg_aqi = merged_multi_city.groupby(['year', 'City'])[AQI_COL].mean().reset_index()

    # Bar plot (existing)
    print("Generating yearly average AQI bar plot...") # Debug print
    if not yearly_avg_aqi.empty:
        plt.figure(figsize=(14, 7))
        sns.barplot(x='year', y=AQI_COL, hue='City', data=yearly_avg_aqi)
        plt.xlabel("Year")
        plt.ylabel(f"Average {AQI_COL}")
        plt.title(f"Yearly Average {AQI_COL} Comparison by City (Bar Plot)")
        plt.legend(title="City", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.savefig(os.path.join(PLOTS_DIR, f"barplot_yearly_avg_{AQI_COL.lower()}_city.png"), dpi=150)
        plt.show()
    else:
        print(f"Yearly average AQI data is empty after aggregation for bar plot. Cannot generate plot.")


    # Line plot (new, based on user request)
    print("Generating yearly average AQI line plot...") # Debug print
    if not yearly_avg_aqi.empty:
        fig_line = px.line(yearly_avg_aqi,
                           x='year',
                           y=AQI_COL,
                           color='City',
                           title=f"Yearly Average {AQI_COL} Trends by City (Line Plot)",
                           labels={'year': 'Year', AQI_COL: f'Average {AQI_COL}'})
        fig_line.update_layout(xaxis = dict(tickmode = 'linear', tick0 = yearly_avg_aqi['year'].min(), dtick = 1)) # Ensure all years are shown
        fig_line.show() # Explicitly show the figure

        # Save the line plot
        png_path_line = os.path.join(PLOTS_DIR, f"lineplot_yearly_avg_{AQI_COL.lower()}_city.png")
        html_path_line = os.path.join(PLOTS_DIR, f"lineplot_yearly_avg_{AQI_COL.lower()}_city.html")
        try:
             fig_line.write_image(png_path_line)
             print(f"Saved line plot to {png_path_line}")
        except Exception as e:
             print(f"Could not save line plot as PNG. Error: {e}")
             print(f"Attempting to save as HTML instead.")
             try:
                  fig_line.write_html(html_path_line)
                  print(f"Saved line plot as HTML to {html_path_line}")
             except Exception as html_e:
                  print(f"Could not save line plot as HTML either. Error: {html_e}")

    else:
        print(f"Yearly average AQI data is empty after aggregation for line plot. Cannot generate plot.")

else:
    print(f"Column {AQI_COL} or 'year' not found for yearly AQI comparison plots.")


# Correlation matrix for main variables (Note: This might be less meaningful for a single matrix over multiple cities unless relationships are consistent)
# Consider calculating correlations per city or on city-aggregated data if needed
corr_cols = [c for c in [PM25_COL, PM10_COL, AQI_COL, "temp_C", "rh_percent", "precip_mm", "wind_m_s"] if c in merged_multi_city.columns] # Added AQI_COL
if len(corr_cols) > 1:
    print("\nCalculating Spearman Correlation Matrix (for merged data - apply with caution for multi-city)...") # Debug print
    # Calculate correlation on data without the 'City' column for the correlation matrix
    corr_data = merged_multi_city[corr_cols].copy().dropna() # Drop NaNs for correlation calculation
    print(f"Correlation matrix data shape: {corr_data.shape}") # Debug print
    if not corr_data.empty:
        corr = corr_data.corr(method="spearman")
        plt.figure(figsize=(8,6))
        sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
        plt.title("Spearman Correlation Matrix (Across all merged data)")
        plt.savefig(os.path.join(PLOTS_DIR, "correlation_matrix_multi_city.png"), dpi=150)
        plt.show()
        print("Correlation (Spearman):")
        display(corr)
    else:
        print("Correlation matrix data is empty after dropping NaNs. Cannot compute correlation.")
else:
    print("Not enough columns for correlation matrix.")


# -----------------------
# STEP 7: Statistical Analysis - Regression (Pollutant ~ weather) - Adapted for Multi-city (Example)
# -----------------------
print("\n--- Starting Statistical Analysis (Step 7) ---") # Debug print
# Note: Regression should ideally be run per city or using city as a categorical variable.
# The current code runs regression on the combined data which might not be appropriate.
# Modifying to run per city as an example:
if PM25_COL in merged_multi_city.columns:
    reg_cols = [c for c in ["temp_C", "rh_percent", "precip_mm", "wind_m_s"] if c in merged_multi_city.columns]
    # Ensure there are weather columns
    if reg_cols:
        print(f"\nPerforming Regression ({PM25_COL} ~ meteorology) per City...")
        regression_summaries = {}
        for city in merged_multi_city['City'].unique():
            print(f"\nRegression for {city}:") # Debug print
            df_reg_city = merged_multi_city[merged_multi_city['City'] == city][[PM25_COL] + reg_cols].dropna()

            if len(df_reg_city) > len(reg_cols) + 1: # Need at least p+1 observations for p predictors
                X_city = df_reg_city[reg_cols]
                y_city = df_reg_city[PM25_COL]
                X_city_const = sm.add_constant(X_city)
                try:
                    model_city = sm.OLS(y_city, X_city_const).fit()
                    print(model_city.summary())
                    regression_summaries[city] = model_city.summary()
                except Exception as e:
                    print(f"Error during regression for {city}: {e}")
            else:
                print(f"Not enough weather data or {PM25_COL} data for regression in {city} after dropping NaNs.")

        # Optionally save summaries
        if regression_summaries:
             with open(os.path.join(PLOTS_DIR, "regression_summaries_per_city.txt"), "w") as f:
                  for city, summary in regression_summaries.items():
                       f.write(f"Regression Summary for {city}\n")
                       f.write("="*30 + "\n")
                       f.write(summary.as_text())
                       f.write("\n\n")
             print(f"Saved regression summaries per city to {os.path.join(PLOTS_DIR, 'regression_summaries_per_city.txt')}")

    else:
        print("No weather data available for regression.")
else:
    print(f"{PM25_COL} not found in dataset; skipping regression.")

# -----------------------
# STEP 8: ML Prediction - Prepare features (lag features + meteorology) - Adapted for Multi-city (Example)
# -----------------------
print("\n--- Starting ML Prediction Preparation (Step 8) ---") # Debug print
# Note: ML prediction should ideally be done per city or using city as a feature.
# The current code trains a single model on the combined data, which might not capture city-specific patterns well.
# We will proceed with a single model on the combined data for simplicity based on the original template structure.
N_LAGS = 7  # number of lag days to use

def create_lag_features_multi_city(df, target_col, n_lags):
    df_feat = df.copy()
    if target_col in df_feat.columns:
        # Apply lag creation within each city group
        for lag in range(1, n_lags+1):
            df_feat[f"{target_col}_lag_{lag}"] = df_feat.groupby("City")[target_col].shift(lag)
    return df_feat

if PM25_COL in merged_multi_city.columns:
    print("\nPreparing ML data with lag features per city...")
    df_ml = merged_multi_city.copy()
    df_ml = create_lag_features_multi_city(df_ml, PM25_COL, N_LAGS)

    # include meteorological columns as features if present
    met_cols = [c for c in ["temp_C", "rh_percent", "precip_mm", "wind_m_s"] if c in df_ml.columns]
    # Include 'City' as a categorical feature (requires one-hot encoding)
    # For simplicity in following the original template structure, we will *not* one-hot encode City here,
    # and just use lag and met features. A more rigorous approach would include City.
    # feature_cols = [c for c in [f"{PM25_COL}_lag_{i}" for i in range(1, N_LAGS+1)] + met_cols if c in df_ml.columns] # Original feature list
    # Let's redefine feature_cols to ensure they exist after lag creation and potential merge
    lag_feature_cols = [f"{PM25_COL}_lag_{i}" for i in range(1, N_LAGS+1)]
    feature_cols = [c for c in lag_feature_cols + met_cols if c in df_ml.columns]


    if not feature_cols:
        raise ValueError("No features available for ML model. Check if PM2.5 and/or weather columns exist and N_LAGS > 0.")

    # Drop rows with NaN in target or feature columns
    df_ml = df_ml[[PM25_COL] + feature_cols].dropna()

    if df_ml.empty:
        print("ML dataset is empty after dropping NaNs. Check data quality and missing values.")
    else:
        print("ML dataset shape:", df_ml.shape)
        display(df_ml.head())

        # Train/test split (time-based - may not be ideal for multi-city if cities have different date ranges)
        # A more robust approach might split per city or use cross-validation.
        # We will stick to a simple time-based split on the combined data for now.
        train_size = int(len(df_ml) * 0.8)
        if train_size == 0 or train_size >= len(df_ml):
             print("Not enough data for train/test split after dropping NaNs.")
        else:
            train_df = df_ml.iloc[:train_size]
            test_df = df_ml.iloc[train_size:]

            X_train = train_df[feature_cols].values
            y_train = train_df[PM25_COL].values
            X_test = test_df[feature_cols].values
            y_test = test_df[PM25_COL].values

            # scale features
            scaler = StandardScaler()
            X_train_s = scaler.fit_transform(X_train)
            X_test_s = scaler.transform(X_test)

            # MLPRegressor
            print("\nTraining MLP Regressor...")
            mlp = MLPRegressor(hidden_layer_sizes=(128,64), max_iter=1000, random_state=RANDOM_SEED)
            mlp.fit(X_train_s, y_train)
            y_pred = mlp.predict(X_test_s)

            print("MLP Results on test set:")
            print("MSE:", mean_squared_error(y_test, y_pred))
            print("R2 :", r2_score(y_test, y_pred))

            # plot sample (Note: This plots across multiple cities mixed in the test set)
            plt.figure(figsize=(12,5))
            # Plotting: Use test_df.index directly if it's a DatetimeIndex
            # If test_df.index might be non-contiguous due to dropped NaNs,
            # consider resetting index for plotting or using a dedicated date column
            plt.plot(test_df.index, y_test, label="Actual PM2.5")
            plt.plot(test_df.index, y_pred, label="Predicted PM2.5", alpha=0.8)
            plt.legend()
            plt.title("MLP: Actual vs Predicted PM2.5 (Test set - Multi-city)")
            plt.savefig(os.path.join(PLOTS_DIR, "mlp_actual_vs_pred_test_multi_city.png"), dpi=150)
            plt.show()

else:
    print(f"{PM25_COL} missing - can't train ML model.")


# -----------------------
# STEP 9: 7-day Forecasting - Adapted for Multi-city (Example)
# -----------------------
print("\n--- Starting 7-day Forecasting (Step 9) ---") # Debug print
# Note: Forecasting with a single model trained on combined data might not be accurate for specific cities.
# For a true multi-city forecast, you might train models per city or include City as a feature.
# The current function forecasts for the last date in the *combined* data, which could be from any city.
# To forecast per city, you would need to iterate through cities and apply the forecast function.
# We will keep the existing forecast function but note its limitation.

def forecast_next_7_days(model, scaler, merged_df, feature_cols, target_col, n_lags=7, future_weather=None):
    """
    Returns a DataFrame with predictions for next 7 days (index = future dates).
    model: trained sklearn regressor (expects scaled features)
    scaler: fitted scaler for features
    merged_df: full historical DataFrame (with daily index)
    feature_cols: list of feature column names used to train model (lags and optionally meteorology)
    target_col: name of target (PM2.5)
    future_weather: DataFrame indexed by future dates with meteorology columns (met_cols) if available
    """
    if merged_df.empty or target_col not in merged_df.columns:
        print("Cannot forecast: Historical data is empty or target column is missing.")
        return pd.DataFrame() # Return empty DataFrame

    last_date = merged_df.index.max()
    preds = {}
    # create a working copy extended as we add predictions
    work = merged_df.copy()

    # Identify meteorological columns from feature_cols
    met_cols_in_features = [c for c in feature_cols if c not in [f"{target_col}_lag_{i}" for i in range(1,n_lags+1)]]

    for day in range(1, 8):
        pred_date = last_date + timedelta(days=day)
        # build feature vector: lags are previous N days of target (either real or predicted)
        feat_dict = {}
        for lag in range(1, n_lags+1):
            lag_col_name = f"{target_col}_lag_{lag}"
            if lag_col_name in feature_cols: # Only add lag feature if it was used in training
                 # Get the value from the working dataframe, shifted appropriately
                 # The shift is based on the *current* state of the working dataframe
                 # which includes historical data and previous predictions
                 # Handle potential NaNs in initial lags by falling back to last known non-NaN value or 0.0
                 lag_val = work[target_col].shift(lag).iloc[-1]
                 if pd.isna(lag_val):
                      # Fallback strategy for missing lags (e.g., at the very beginning of the forecast period)
                      # Simple fallback to the last valid value, or 0 if no valid value found
                      last_valid_hist_value = work[target_col].dropna().iloc[-1] if not work[target_col].dropna().empty else 0.0
                      lag_val = last_valid_hist_value # Using last valid historical value as a simple fallback

                 feat_dict[lag_col_name] = lag_val


        # meteorology values for pred_date
        met_vals = {}
        for col in met_cols_in_features:
            if future_weather is not None and pred_date in future_weather.index and col in future_weather.columns:
                met_vals[col] = future_weather.loc[pred_date, col]
            else:
                # fallback: use last available meteorology value (persistence) or 0.0
                if col in work.columns and not work[col].isna().iloc[-1]:
                    met_vals[col] = work[col].iloc[-1]
                else:
                    # Use mean from training data for missing future weather, or 0.0 if no training data mean
                    # Requires access to training data or its stats. For simplicity, using 0.0 or last known.
                    # A more robust approach would use training set mean/median for imputation.
                    met_vals[col] = 0.0 # Using 0.0 as a simple fallback
            feat_dict[col] = met_vals[col] # Add met value to feature dict

        # Ensure feature vector has all columns used during training, fill missing with 0 or mean if needed
        # This step is crucial if the feature_cols list includes items not always available (like precip_mm)
        # For simplicity here, we'll assume feature_cols match what's built in feat_dict if data is available.
        # A more robust approach would ensure the order and presence of all feature_cols.
        # Let's rebuild the feature array based on the exact order of feature_cols used in training
        x_arr = np.array([feat_dict.get(col, 0.0) for col in feature_cols]).reshape(1, -1) # Use 0.0 as fallback

        try:
            x_s = scaler.transform(x_arr)
            pred_val = model.predict(x_s)[0]
            preds[pred_date] = pred_val

            # append predicted value to work so next iteration uses it as lag
            # Create a new row with the prediction and any available met data for the prediction date
            new_row_data = {target_col: pred_val}
            for met_col, val in met_vals.items():
                 new_row_data[met_col] = val # Add met values to the new row

            new_row_df = pd.DataFrame(new_row_data, index=[pred_date])
            work = pd.concat([work, new_row_df]) # Use concat for appending

        except Exception as e:
             print(f"Error during forecasting for date {pred_date}: {e}")
             preds[pred_date] = np.nan # Mark prediction as NaN on error
             # Still append a row to work to maintain index structure, even if prediction is NaN
             new_row_data = {target_col: np.nan}
             for met_col, val in met_vals.items():
                  new_row_data[met_col] = val
             new_row_df = pd.DataFrame(new_row_data, index=[pred_date])
             work = pd.concat([work, new_row_df])


    preds_df = pd.Series(preds).to_frame(name=f"{target_col}_pred")
    return preds_df

# try to build a future_weather DataFrame if you have future weather data (USER supply). Here we set None.
# Example: future_weather = pd.read_csv("future_weather.csv", index_col="date", parse_dates=True)
future_weather = None  # <-- If you have future weather CSV, load it here (index=dates) with temp_C,rh_percent,precip_mm,wind_m_s columns.

# Check if MLP model was successfully trained before forecasting
if 'mlp' in locals() and 'scaler' in locals() and 'feature_cols' in locals() and PM25_COL in merged_multi_city.columns and not merged_multi_city.empty:
    print("\nAttempting 7-day forecasting...")
    # Ensure df_ml_for_forecast is created based on the full merged_multi_city data
    # and includes the 'City' column for lag calculation within groups
    df_ml_for_forecast = create_lag_features_multi_city(merged_multi_city.copy(), PM25_COL, N_LAGS)

    # Ensure df_ml_for_forecast has the columns expected by feature_cols + City
    required_cols_df_ml = [PM25_COL] + [f"{PM25_COL}_lag_{i}" for i in range(1, N_LAGS+1)] + met_cols + ['City']
    for col in required_cols_df_ml:
         if col not in df_ml_for_forecast.columns:
              print(f"Warning: Column '{col}' missing in forecast data. Adding with NaNs.")
              df_ml_for_forecast[col] = np.nan # Add missing columns


    # Filter to the latest data point for each city to start forecasting from there.
    # This requires iterating through cities to forecast.
    print("Forecasting for the last date available in each city...")
    all_city_preds = []
    for city in merged_multi_city['City'].unique():
        print(f"\nForecasting for {city}...")
        # Get the last N_LAGS data points *for this city* plus any needed met data
        df_city_latest = df_ml_for_forecast[df_ml_for_forecast['City'] == city].copy()
        if len(df_city_latest) > N_LAGS: # Need enough data to create initial lags
            # The forecast_next_7_days function expects the full historical data up to the forecast start.
            # Pass the data for the specific city to the forecast function.
            preds_df_city = forecast_next_7_days(mlp, scaler, df_city_latest, feature_cols, PM25_COL, n_lags=N_LAGS, future_weather=None) # future_weather is None for this example

            if not preds_df_city.empty:
                 preds_df_city['City'] = city # Add City column to predictions
                 all_city_preds.append(preds_df_city)
            else:
                 print(f"7-day forecasting failed for {city}.")
        else:
             print(f"Not enough data for {city} to generate lags for forecasting ({len(df_city_latest)} days available, need > {N_LAGS}). Skipping forecast for this city.")


    if all_city_preds:
        preds_df = pd.concat(all_city_preds) # Combine predictions from all cities
        print("\nCombined 7-day predictions (PM2.5) for all cities:")
        display(preds_df)

        # Plot predictions appended to historical series *per city*
        print("\nGenerating forecast plots per city...")
        for city in preds_df['City'].unique():
             plt.figure(figsize=(12,5))
             hist_data_city = merged_multi_city[merged_multi_city['City'] == city]
             preds_data_city = preds_df[preds_df['City'] == city]

             # Plot last 60 days of historical data for this city
             if len(hist_data_city) > 60:
                 plt.plot(hist_data_city.index[-60:], hist_data_city[PM25_COL].iloc[-60:], label=f"Historical {PM25_COL} ({city})")
             else:
                 plt.plot(hist_data_city.index, hist_data_city[PM25_COL], label=f"Historical {PM25_COL} ({city})")

             plt.plot(preds_data_city.index, preds_data_city[f"{PM25_COL}_pred"], marker="o", linestyle="--", label=f"7-day Forecast ({city})")
             plt.title(f"7-day Forecast of {PM25_COL} for {city} (recursive)")
             plt.xlabel("Date")
             plt.ylabel(PM25_COL)
             plt.legend()
             plt.savefig(os.path.join(PLOTS_DIR, f"7day_forecast_{PM25_COL.lower()}_{city.lower().replace(' ', '_')}.png"), dpi=150)
             plt.show()

        # Save combined predictions to CSV
        preds_df.to_csv("pm25_7day_forecast_multi_city.csv") # Changed output name
        print("Saved combined 7-day forecast to pm25_7day_forecast_multi_city.csv")

    else:
        print("7-day forecasting skipped as no predictions were generated for any city.")

else:
    print("Skipping 7-day forecasting: ML model was not successfully trained or data is missing.")


# -----------------------
# STEP 10: Save model outputs & artifacts - Adapted for Multi-city
# -----------------------
# Save combined predictions to CSV (already done in STEP 9)
# if 'preds_df' in locals() and not preds_df.empty:
#     preds_df.to_csv("pm25_7day_forecast_multi_city.csv")
#     print("Saved combined 7-day forecast to pm25_7day_forecast_multi_city.csv")
# else:
#     print("No combined 7-day forecast to save.")

# Save plots already saved in PLOTS_DIR (PNG and/or HTML)

# Save sample correlation/regression tables to CSV
if 'corr' in locals():
    corr.to_csv("correlation_matrix_spearman_multi_city.csv") # Changed output name
    print("Saved correlation matrix (multi-city) to correlation_matrix_spearman_multi_city.csv")
else:
    print("No correlation matrix to save.")

# Regression summaries per city saved in STEP 7 if generated.

# -----------------------
# STEP 11: Quick Summary (print key numbers) - Adapted for Multi-city
# -----------------------
print("\n--- Quick Summary (Multi-city) ---")

# Summarize regression results per city
if 'regression_summaries' in locals() and regression_summaries:
    print(f"Regression ({PM25_COL} ~ meteorology) Adjusted R^2 per City:")
    for city, summary in regression_summaries.items():
        try:
            adj_r2 = summary.rsquared_adj
            print(f" - {city}: adj. R^2 = {adj_r2:.3f}")
        except:
            print(f" - {city}: Could not compute adjusted R^2.")
else:
     print(f"Regression models for {PM25_COL} were not created for any city.")


# Summarize MLP test metrics (Note: These are for the single model on combined data)
if 'y_test' in locals() and 'y_pred' in locals():
    try:
        print(f"\nMLP test metrics (single model on combined data):")
        print(f"MSE: {mean_squared_error(y_test, y_pred):.3f}")
        print(f"R2 : {r2_score(y_test, y_pred):.3f}")
    except:
        print("\nCould not compute MLP test metrics.")
else:
    print("MLP test metrics not available (model not trained or test set empty).")

print("\nTop-level files potentially saved:")
print(" - cleaned dataset:", CLEANED_CSV_OUT)
print(" - plots folder:", PLOTS_DIR)
print(" - combined 7-day forecast:", "pm25_7day_forecast_multi_city.csv (if generated)")
print(" - correlation matrix:", "correlation_matrix_spearman_multi_city.csv (if generated)")
print(" - regression summaries per city:", f"{PLOTS_DIR}/regression_summaries_per_city.txt (if generated)")


# -----------------------
# NOTES & Next Steps
# -----------------------
# - Review the output plots and summaries for each city.
# - Check the column names used (DATE_COL, PM25_COL, etc.) in STEP 2 and adjust if needed based on your CSV files' actual column headers.
# - If the date column or pollutant columns were not found for any city, inspect the output to identify the correct names and update STEP 2.
# - The NASA POWER fetch is currently for a single point and applied to all cities. For more accurate analysis, consider obtaining city-specific weather data.
# - The ML model and forecasting are currently performed on the combined multi-city dataset. For potentially better performance, consider training separate models for each city or including 'City' as a categorical feature in the model.
# - Consider refining data cleaning (imputation), feature engineering (e.g., holidays, events), and ML models/hyperparameters.
# - The comparative yearly and monthly plots provide key insights into city-level trends and seasonality.
# - Use the generated visuals, summary statistics, and forecast results to build your project report and presentation.
```
